```python
import pandas as pd

# Cargar datos de ventas
ventas_df = pd.read_csv('datos/ventas (1).csv')

# Cargar datos de clientes
clientes_df = pd.read_json('datos/clientes.json')

# Cargar datos de inventario
inventario_df = pd.read_excel('datos/inventario.xlsx')
```


**Integración de datos:**

Se intentó integrar los datos de ventas y clientes mediante la función `merge()` de Pandas.

Sin embargo, no fue posible realizar la combinación de los DataFrames debido a que no existe una columna en común entre ambos conjuntos de datos.

El DataFrame de ventas contiene la columna "id_venta", mientras que el de clientes contiene "id_cliente", sin una relación directa entre ellos.

Por este motivo, no se pudo realizar la integración de datos en este caso.
